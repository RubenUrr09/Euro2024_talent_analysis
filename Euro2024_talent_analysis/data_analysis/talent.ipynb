##  EURO 2024 Talent Analysis
Purpose for analyzing this dataset:
The EURO 2024 is one of the most important tournaments in the sport - Football. Considered in the top 3 most important tournaments being:
1. FIFA World Cup
2. EURO 2024/Copa America 2024
3. UEFA Champions League

So, why are we analyzing this data for the EURO 2024? 
Since this is one of the most important tournaments in football, it is the best chance for young talent to get their names to be known all aorund the world.
To find the best talents in this tournament, we'll analyze players under 23 years of age with:

Forwards and Midfielders
1. Goals scored: 
Determines the players that contributed the most to the goals for their national team. Better suited for forwards

2. Shots created and shooting:
Determines players that have great positioning and confidence to shoot the ball. Better suited for forwards or attacking midfielders

3. Passing:
Determines players that have great passing ability and playmaking. Better suited for midfielders but still a great asset for forwards

4. Posession:
Determines how long the players kept the ball, successful dribbles

Defenders:
1. Tackles, headers/duels won:
Determines success of stealing the ball, or heading the ball

2. Passing:
Determines players that have great passing ability and playmaking. Better suited for midfielders but still a great asset for forwards

3. Posession:
Determines how long the players kept the ball, successful dribbles

# Forwards / Midfielders

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
string = os.getcwd() 
string = string.replace('data_analysis', 'datasets')
```

```python
goal_and_shot_creation_path = string + r'\\player_goal_and_shot_creation.csv'
goal_shot_c = pd.read_csv(goal_and_shot_creation_path)
goal_shot_c
```

## SCA: Shot Creating Actions
The two offensive actions directly leading to a shot, such as passes, take-ons and drawing fouls

```python
goal_shot_c.columns
```

```python
shots = goal_shot_c[['Player', 'Age', 'SCA']]
shots['Age'] <= 23
young_shots = shots[ shots['Age'] <= 23]
young_shots
```

```python
plt.plot(young_shots['Player'], young_shots['SCA'])
```

```python
young_shots = young_shots.sort_values(by=['SCA'])
```

```python
young_shots.tail(20)
```

Among these top 20 players based on their shot creation contributions, there are some names that are already known worldwide such as Lamine Yamal, Nico Williams, Xavi Simons, Bukayo Saka, etc. 
Something notice from this top 20 contributors are names that could rise up later such as: Kenan Yildiz (Having a great season at the club Juventus by the end of 2024), Kristjan Asllani, Francisco Conceição, Heorhiy Sudakov, Barış Alper Yılmaz, Luka Sučić

# Defenders
We'll analyze tackles, aerial duels won, headers and posession

```python
defenderspath = string + r'\\player_defense_actions.csv'
defensestats = pd.read_csv(defenderspath)
```

```python
defensestats.columns
```

## Tackles Won

```python
young_tackles_won = defensestats[['Player', 'Pos', 'Age', 'TklW']]
young_tackles_won = young_tackles_won[young_tackles_won['Age'] <= 23]
young_tackles_won = young_tackles_won[young_tackles_won['Pos'].str.contains('DF', na=False)]
young_tackles_won
```

Something we can immediately notice is the lack of young defenders in the tournament. This is due to their lack of experience playing the sport. Coaches would prefer to play defenders that have played the sport for longer time because they get less nervous when playing against world-class teams, and can communicate more effectively due to their experience. They more experienced changing the tactics of their defense based on how the other team is playing, too.

We'll analyze these young talents that have earned their spots over defenders of more age/experience

```python
young_tackles_won = young_tackles_won.sort_values(by=['TklW'])
young_tackles_won
```

We can see that among the defenders with the most successful tackles are players such as: Josip Stanišić, Strahinja Pavlović, Nuno Mendes, Bukayo Saka, and others
